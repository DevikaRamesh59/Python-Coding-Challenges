### **Student Grade and Hobbies**

### **Dictionary, Set, Frozen Set**

```python
from typing_extensions import FrozenSet
### Create a dictionary with students name as keys and there grades as values
Dict_Stu={"Raja": 3,"Saai": 1,"Devika":4,"Magna":2}

## Allow the users to add or update the grades for at least three students
Dict_Stu["Saai"]=1
Dict_Stu["Devika"]=2
Dict_Stu["Magna"]=3
Dict_Stu
```

```python
###Create a set of hobbies and allow the users to input atleast 3 hobbies.
Hobbies=set()
Hobbies1=input("Enter hobby 1")
Hobbies.add(Hobbies1)
Hobbies2=input("Enter hobby 2")
Hobbies.add(Hobbies2)
Hobbies3=input("Enter hobby 3")
Hobbies.add(Hobbies3)
print(Hobbies)
```

```python
## Convert the set of hobbies into frozen set and display it.
Fr_Hobbies=frozenset(Hobbies)
print(Fr_Hobbies)
```
